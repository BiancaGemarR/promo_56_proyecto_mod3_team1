```python
import csv 
import json 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("hr_clean_data_2023.csv")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='age', hue='joblevel', palette='magma', ci=None)
plt.xticks(rotation=90)
plt.title('Edad y cargos')
plt.ylabel('Edad') 
plt.xlabel('Cargo en la empresa')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='remotework', hue='jobrole', palette='magma')

plt.title('Teletrabajo y cargos')
plt.ylabel('Teletrabajo') 
plt.xlabel('Cargo')
plt.legend(title='Cargo')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='jobrole', hue='businesstravel', palette='inferno')
plt.xticks(rotation=90)
plt.title('Viajes en la empresa')
plt.ylabel('Cargo') 
plt.xlabel('Viajes')
plt.legend(title='Frecuencia de viaje')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='joblevel', y='monthlyincome', hue='education', ci=None)
plt.title('Salario mensual promedio por nivel de trabajo y educación')
plt.xlabel('Nivel de trabajo')
plt.ylabel('Salario mensual promedio')
plt.legend(title='Nivel de educación')
plt.show()
```

```python
#years in current role tiene muchos nulos igual no tiene sentido
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='yearsatcompany', hue='yearsincurrentrole', palette='PuOr',ci=None)
plt.xticks(rotation=90)
plt.title('Años en la compañía y cargo')
plt.ylabel('Años en la compañía') 
plt.xlabel('Años en el cargo')
plt.legend(title='Años en el cargo')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='percentsalaryhike', hue='jobsatisfaction', palette='Dark2')
plt.xticks(rotation=90)
plt.title('Subida del salario y satisfacción')
plt.ylabel('Subida del salario') 
plt.xlabel('Satisfacción en el trabajo')
sns.despine(top=True, right=True)
```
