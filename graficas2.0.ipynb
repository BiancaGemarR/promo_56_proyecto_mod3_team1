```python
import csv 
import json 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("hr_clean_data_2023.csv")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='age', hue='joblevel', palette='magma')
plt.xticks(rotation=90)
plt.title('Edad y cargos')
plt.ylabel('Edad') 
plt.xlabel('Cargo en la empresa')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='yearsatcompany', hue='yearsincurrentrole', palette='plasma')
plt.xticks(rotation=90)
plt.title('Aos en la compaa y cargo')
plt.ylabel('Aos en la compaa') 
plt.xlabel('Aos en el cargo')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='percentsalaryhike', hue='relationshipsatisfaction', palette='inferno')
plt.xticks(rotation=90)
plt.title('Subida del salario y satisfaccin')
plt.ylabel('Subida del salario') 
plt.xlabel('Satisfaccin en el entorno')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='remotework', hue='jobrole', palette='viridis')
plt.xticks(rotation=90)
plt.title('Teletrabajo y cargos')
plt.ylabel('Teletrabajo') 
plt.xlabel('Cargo')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, y='jobrole', hue='businesstravel', palette='magma')
plt.xticks(rotation=90)
plt.title('Viajes en la empresa')
plt.ylabel('Cargo') 
plt.xlabel('Viajes')
sns.despine(top=True, right=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='stockoptionlevel', hue='joblevel', palette='magma')
plt.xticks(rotation=90)
plt.title('Acciones que da la empresa')
plt.ylabel('Nivel de acciones') 
plt.xlabel('Nivel del trabajo')
sns.despine(top=True, right=True)
```
